###  I want to know the dependency between  Features of mobile and Price of mobile by building Machine leaarning model hence to know the exact price of new mobile to buy. For this purpose I will use Web-scraping to collect data. 

```python
#Importing Libraries 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
#Url for scraping
url=requests.get('https://www.flipkart.com/search?q=iphone&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_0_6&otracker1=AS_Query_HistoryAutoSuggest_0_6&as-pos=0&as-type=HISTORY&as-searchtext=iphone')
```

```python
#Printing part of html
print(url.text[0:500])
```

```python
#Parsing a page with BeautifulSoup

soup=BeautifulSoup(url.text, 'html.parser')
```

```python
#Use function “prettify” to look at nested structure of HTML page
print(soup.prettify())
```

```python
#To get links of all items
results=soup.find_all("div", {"class":"_3O0U0u"})
```

```python
#Checking for no.of items
len(results)
```

```python
#Looking into first item of the page
first_result=results[1]
print(first_result)
```

```python
#To get Name tag
name=first_result.findAll('div', {'class': '_3wU53n'})
name=name[0].text
name
```

```python
#To get RAM
Ram=name.split(' ')
Ram=Ram[-2]+Ram[-1]
Ram[:-1]
```

```python
#To get colour
colour=name.split(' ')
colour=colour[-3][:-1]
colour
```

```python
#To get Price tag
price= first_result.findAll('div', {'class': '_1vC4OE _2rQ-NK'})
price[0].text
```

```python
#To get Rating tag
rating=first_result.findAll('div', {'class': 'hGSR34'})
rating[0].text
```

```python
#To get details of all items

records=[]
for result in results:
    name=result.findAll('div', {'class': '_3wU53n'})
    name=name[0].text
    
    RAM=name.split(' ')
    RAM=(RAM[-2]+RAM[-1])[:-1]
    
    colour=name.split(' ')
    colour=colour[-3][:-1]

    price= result.findAll('div', {'class': '_1vC4OE _2rQ-NK'})
    price=price[0].text
    
    rating=result.findAll('div', {'class': 'hGSR34'})
    rating=rating[0].text
    
    specifications=result.findAll('li', {'class': 'tVe95H'})
    ROM=specifications[0].text
    Display=specifications[1].text
    Camera=specifications[2].text
    Processor=specifications[3].text
    Warranty=specifications[4].text
    
    
    records.append((name, RAM, colour, rating, ROM, Display, Camera, Processor, Warranty, price))
```

```python
records[0]
```

```python
len(records)
```

### conversion of data into Dataframe

```python
Data=pd.DataFrame(records, columns=['Mobile', 'RAM', 'Colour', 'Rating', 'ROM', 'Display', 'Camera', 'Processor', 'Warranty', 'Price'])
```

```python
Data.head()
```

Data Preprocessing for model building

```python
Dataset=Data
```

```python
Dataset.isna().sum()
```

```python
Dataset.dtypes
```

```python
Dataset['RAM']=Dataset['RAM'].apply(lambda x:x[:-2])
Ram_dummies=pd.get_dummies(Dataset['RAM'], prefix='RAM')
Dataset=pd.concat([Dataset, Ram_dummies],  axis=1)
Dataset.drop('RAM',axis=1, inplace=True)
```

```python
Dataset['Colour']=Dataset['Colour'].apply(lambda x:x if x[0]!='(' else x[1:])
colour_dummies=pd.get_dummies(Dataset['Colour'], prefix='colour')
Dataset=pd.concat([Dataset, colour_dummies],  axis=1)
Dataset.drop('Colour',axis=1, inplace=True)
```

```python
Dataset['Rating']=Dataset['Rating'].astype('float')
```

```python
Dataset['Display']=Dataset['Display'].apply(lambda x:x[:5])
Dataset['Display']=Dataset['Display'].astype('float')
```

```python
Dataset['ROM']=Dataset['ROM'].apply(lambda x:x[0:-9])
Dataset['ROM']=Dataset['ROM'].astype('int')
```

```python
Dataset['Price']=Dataset['Price'].apply(lambda x:x[1:])
Dataset['Price']=Dataset['Price'].apply(lambda x:x.split(',')[0]+x.split(',')[1])
Dataset['Price']=Dataset['Price'].astype('int')
```

```python
Train=Dataset.drop(['Mobile', 'Camera', 'Processor', 'Warranty'], axis=1)
```

### Linear-Regression model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=Train.drop('Price', axis=1)
y=Train.Price
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
Model=LinearRegression()
```

```python
Model.fit(x_train, y_train)
```

```python
Prediction=Model.predict(x_test)
```
